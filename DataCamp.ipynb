#  Creating a path for Data importing and converting into csv.file

```python

import pandas as pd
import numpy as np

np.random.seed(42)

sales_methods = ["Email", "Call", "Email and Call"]
states = ["California", "New York", "Texas", "Florida", "Illinois"]
weeks = np.arange(1, 13)  # 12 weeks of data

num_records = 500
data = {
    "week": np.random.choice(weeks, num_records),
    "sales_method": np.random.choice(sales_methods, num_records),
    "customer_id": [f"CUST_{i}" for i in range(1, num_records + 1)],
    "nb_sold": np.random.randint(1, 20, num_records),
    "revenue": np.round(np.random.uniform(50, 1000, num_records), 2),
    "years_as_customer": np.random.randint(1, 40, num_records),
    "nb_site_visits": np.random.randint(0, 50, num_records),
    "state": np.random.choice(states, num_records)
}
df = pd.DataFrame(data)
output_path = "product_sales_sample.csv"
df.to_csv(output_path, index=False)

print(f"File saved successfully at: {output_path}")
```

```python
import os
output_path = os.path.join(os.getcwd(), "product_sales_sample.csv")
df.to_csv(output_path, index=False)
print(f"File saved successfully at: {output_path}")
```

# Exploratory Data Analysis in Python 

Customer Count by Sales Method:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=data, x='sales_method', palette='Set2')
plt.title("Customer Count by Sales Method")
plt.show()
```

```python
# Revenue Distribution:
sns.histplot(data['revenue'], kde=True, color='skyblue')
plt.title("Revenue Distribution")
plt.show()
```

```python
# Weekly Revenue Trend:
sns.lineplot(data=data, x='week', y='revenue', hue='sales_method')
plt.title("Weekly Revenue Trend by Sales Method")
plt.show()
```
